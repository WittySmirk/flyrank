## 3. Quick look at the data (2-3 real numbers)


```python
import os
import pandas as pd, numpy as np

if os.path.basename(os.getcwd()) in ("notebooks", "work"):
    os.chdir("..")
assert os.path.exists("data/raw/content_refresh_anonymized.csv"), "starter CSV not found — are you at the repo root?"

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df_ranked = df[df["avg_position"] > 0]

corr = df_ranked["avg_position"].corr(df_ranked["impressions_90d"])
print("correlation between avg_position and impressions_90d: ", corr)

print("ctr based on fresh_ness_tier: ", df.groupby("freshness_tier")["ctr"].agg(["mean", "median", "count"]))

down = df[df["trend_direction"] == "down"]
print(f"median impressions_90d with downward trend: ", down["impressions_90d"].median())
print(f"median days_since_last_update with downward trend: ", down["days_since_last_update"].median())
```
